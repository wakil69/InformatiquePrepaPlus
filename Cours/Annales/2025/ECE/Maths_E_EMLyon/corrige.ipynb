## Exercice 2 : Partie B - Question 9

```python
def voisins(A, i):
    n = len(A[i])
    V = []
    for j in range(n):
        if j != i and A[i, j] != 0:
            V.append(j)
    return V
```

## Exercice 2 : Partie B - Question 10

```python
def min_ext(L):
    m = 0
    while m in L:
        m = m + 1
    return m
```

## Exercice 2 : Partie B - Question 10

```python
def coloration(A):
    n = len(A[0])
    C = [k for k in range(0, n)]
    for i in range(1, n):
        C_voisins = [C[j] for j in voisins(A, i)]
        C[i] = min_ext(C_voisins)
    return C
```
