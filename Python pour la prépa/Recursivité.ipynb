**Commencer par de petits cas vous permet de trouver la formule générale**

```python
Image.open("5!.png")
```

```python
def fact(n):
    if n==0 or n==1:
        return 1
    else:
        return n*fact(n-1)
```

```python
print(5*4*3*2*1)
print(fact(5))
```

```python
def reverse(word):
    if len(word) in [0,1]:
        return word
    else:
        return word[-1] + reverse(word[:-1])
```

```python
print(reverse("abcde"))
word = "abcde"
print(word[::-1])
```
